# K means

```python
# prediction and train & test is not posible
#k is no. of cluster we needed
#and k mean is iterative process
#centroid is selected randomly at starting
#by compairing all values , we assign values to that pertucular cluster (with the help of less difference)
#then calculate the mean of cluster and start the next iteration
#and at the end difference between the centroids is quit less 
```

```python
# optimum value of k
#value of is calculated by elbow method(on x axis has value of k & on y axis we have sum square distance)
# wherever i see a sudeen drop in distance i take that correspondin value as optimum value of k
```

```python
# coding
import pandas as pd
ctg = pd.read_csv(r"F:\python\dataset\CTG.csv")
ctg
```

```python
ctg=ctg.drop('NSP',axis=1)
ctg
```

```python
from sklearn.cluster import KMeans
```

```python
ctg_kmeane = KMeans(n_clusters=2)
ctg_kmeane.fit(ctg)
```

```python
# see centroid
ctg_kmeane.cluster_centers_
```

```python
ctg_kmeane.labels_
```

```python
ctg['ctg_labels']=ctg_kmeane.labels_
ctg
```

```python
ctg.sort_values(['ctg_labels'])
```

```python
ctg.ctg_labels.value_counts()
```

```python
ctg_kmeane.fit(ctg).score(ctg) # this it total square distance
```

```python
# it is also called non deterministics algorithem
#as starting centroid changes because it is selected randomly
```

```python
# hyper parameter in kmeans is k (as it is not part of youe data)
```

```python
list1 = list()
num = list(range(1,7))
for i in num:
    ctg_kmeane = KMeans(n_clusters=i)
    ctg_kmeane.fit(ctg)
    ctg_kmeane.fit(ctg).score(ctg)
    list1.append(ctg_kmeane.fit(ctg).score(ctg)*(-1))
    
```

```python
import matplotlib.pyplot as plt
```
